```python
n = int(input())
dist = [0 for _ in range(n)]
arr = []
arr = [int(i) for i in input().split()]

for i in range(0, n):
    a, b = 0, 0
    if (i > 0 and arr[i - 1] < arr[i]):
        a = dist[i - 1] + 1
    if (i < n - 1 and arr[i] > arr[i + 1]):
        j = i
        while (j <= n - 2 and arr[j] > arr[j + 1]):
            b += 1
            j += 1
    dist[i] = max(a, b)

dist[n - 1] = dist[n - 2] + 1 if arr[n - 1] > arr[n - 2] else 0
print(n + sum(dist))
```
