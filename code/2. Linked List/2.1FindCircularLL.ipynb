### 141. Linked List Cycle
Link: https://leetcode.com/problems/linked-list-cycle/description/

#### My ideas:
Link: https://leetcode.com/problems/linked-list-cycle/submissions/1204419130/
- Floyd's Cycle-Finding: `fast` (2 steps) and `slow` (1 step), if there is cycle eventually **2 pointer met (Physic 12)**


```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def hasCycle(self, head: ListNode) -> bool:
        slow = head
        fast = head            
        while fast.next:
            if fast.next.next is None: return False
            slow = slow.next
            fast = fast.next.next
            if slow == fast: return True
        return False
```
