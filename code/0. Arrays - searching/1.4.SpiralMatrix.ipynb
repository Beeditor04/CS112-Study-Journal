### 59. Spiral Matrix II
Link: https://leetcode.com/problems/spiral-matrix-ii/solutions/4890821/python-simple-approach-using-4-direction-list-for-simplicity-beat-97/

### My ideas: 
> Treat it as circle, starting point at `[0 + i][0 + i]` \
where `i` in `[0; 2n - 1]`
- if `n` input is even, there's alway a **center in circle**, just solve it which `if` at the end



```python
class Solution:
    def generateMatrix(self, n: int) -> list[list[int]]:
        count = 1
        matrix = [[0 for _ in range(n)] for _ in range(n)]
        x = [1, 0, -1, 0]  #right, down, left, up
        y = [0, 1, 0, -1]
        indexX = 0
        indexY = 0
        while (n - 1):
            for i in range((n-1)*4 - 1):
                for _ in range(n - 1):
                    matrix[indexX][indexY] = count
                    count = count + 1
                    indexX = indexX + x[i]
                    indexY = indexY + y[i]
            n -= 1
        return matrix
    

ans = Solution()
ans.generateMatrix(3)
```
